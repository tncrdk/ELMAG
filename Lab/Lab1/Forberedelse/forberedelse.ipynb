```python
import numpy as np
import matplotlib.pyplot as plt
```

### Constants

```python
my_0 = 4 * np.pi * 1e-7
N = 330
R = 70e-3
I = 1

max_x = 20E-2
min_x = -20E-2
number_of_data_points = 1000

x_values = np.linspace(min_x, max_x, number_of_data_points)
a_values = [R/2, R, 2*R]
```

### Oppg. 1

```python

def B(x: np.ndarray) -> np.ndarray:
    return N * my_0 * I / (2 * R) * (1 + x**2 / R**2) ** (-3 / 2)
```

```python
B_values = B(x_values)
```

```python
def plot_data(x_values: np.ndarray, y_values_list: list[np.ndarray], plot_name: str) -> None:
    for y_values in y_values_list:
        plt.plot(x_values, y_values)
    plt.savefig(plot_name)
```

```python
plot_data(x_values, [B_values], "oppg_1.png")
```

### Helmholtzspoler

```python
def Helmholtzs_B(x: np.ndarray, a: float) -> np.ndarray:
    return B(x-a/2) + B(x+a/2)
```

```python
B_values_list = [Helmholtzs_B(x_values, a) for a in a_values]
plot_data(x_values, B_values_list, "Helmholtzspole.png")
```

### Anti-Helmholtzspoler

```python
def anti_helmholtz_B(x: np.ndarray, a: float) -> np.ndarray:
    return B(x-a/2) - B(x+a/2)
```

```python
B_values_list = [anti_helmholtz_B(x_values, a) for a in a_values]
plot_data(x_values, B_values_list, "Anti-Helmholtzspole.png")
```
